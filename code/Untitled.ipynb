```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0,8,0.4)
print(len(X))
def f(x):
    y = 2*np.sin(x)+2
    return y
Y = np.zeros(len(X))
for i in range(len(X)):
    Y[i]=f(X[i])
print(Y.shape)
plt.figure(figsize=(15,5))
plt.plot(X,Y)

meas = Y + np.random.normal(0.,0.5,len(X))
plt.plot(X,meas,'o')
```
